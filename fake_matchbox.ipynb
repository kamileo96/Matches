```python
import numpy as np, matplotlib.pyplot as plt
```

```python
def myexp(x, a, b, c, x0):
        omega = 500
        s = 0
        for i in 2**np.arange(6):
                s += np.cos(i*c*(x-x0))/i
        return a*(s)*np.exp(-b*abs(x-x0))*(x>x0) + a*(1/2 + np.cos(omega*(x-x0))/2)*(x>(x0-np.pi/omega))*(x<x0)
```

```python
basex = np.linspace(0,1.5,10000)
a,b,c = 1,8,3000
plt.plot(basex, myexp(basex, a, b, c,0.5))
```

```python
rez = np.random.uniform(-1,1,len(basex))
n = 400
for i in range(n):
    rez += myexp(basex, np.random.uniform(-1,1),4,100,np.random.normal(0.5, 0.2))
```

```python
plt.plot(basex,rez)
```
